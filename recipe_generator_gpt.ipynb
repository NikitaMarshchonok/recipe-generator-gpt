```python
import pandas as pd
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    TrainingArguments,
    Trainer
)
```

```python
df = pd.read_csv("mini_dataset.csv")

df.head()
```

```python
dataset = Dataset.from_pandas(df[['ingredients']].rename(columns={"ingredients": "text"}))

# Convert text from a list string back to a string
dataset = dataset.map(lambda x: {"text": " ".join(eval(x["text"]))})
dataset = dataset.select(range(100))  # ⚡ Только 100 примеров для ускорения
```

```python
tokenizer = AutoTokenizer.from_pretrained("distilgpt2")
model = AutoModelForCausalLM.from_pretrained("distilgpt2")
```

```python
# Add pad_token (otherwise there will be an error when padding=“max_length”)
# tokenizer.pad_token = tokenizer.eos_token

def tokenize(example):
    out = tokenizer(example["text"], truncation=True, padding="max_length", max_length=128)
    out["labels"] = out["input_ids"].copy()
    return out

tokenized_dataset = dataset.map(tokenize, batched=True)

```

```python
from transformers import TrainingArguments, Trainer

# Specify training parameters
training_args = TrainingArguments(
    output_dir="./distilgpt2_recipe_model",   # Where the model is saved
    per_device_train_batch_size=4,            # Batch size
    num_train_epochs=3,                       # Number of epoch
    logging_steps=100,                        # Logging frequency
    save_steps=500,                           # Frequency of model retention
    save_total_limit=2,                       # Only the last 2 checkpoints are saved
    report_to="none"                          # Disable loggers (WandB, etc.)
)

# create Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    tokenizer=tokenizer
)
```

```python
# let's start training
trainer.train()
```

```python
prompt = (
    "To bake a soft and moist chocolate cake, you will need the following ingredients "
    "and preparation steps. First, gather the ingredients:"
)

tokens = tokenizer(prompt, return_tensors="pt", padding=True).to(model.device)

output = model.generate(
    input_ids=tokens["input_ids"],
    attention_mask=tokens["attention_mask"],
    max_length=250,
    temperature=1.05,
    top_k=80,
    top_p=0.95,
    repetition_penalty=1.2,
    num_return_sequences=1,
    do_sample=True,
    pad_token_id=tokenizer.eos_token_id
)

print(tokenizer.decode(output[0], skip_special_tokens=True))
```

```python
# 🟢 Cell 11: Install and create Gradio demo
!pip install -q gradio

import gradio as gr

# Define generation function
def generate_recipe(prompt):
    inputs = tokenizer(prompt, return_tensors="pt", padding=True).to(model.device)
    outputs = model.generate(
        input_ids=inputs["input_ids"],
        attention_mask=inputs["attention_mask"],
        max_length=250,
        temperature=1.05,
        top_k=80,
        top_p=0.95,
        repetition_penalty=1.2,
        num_return_sequences=1,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Create Gradio interface
demo = gr.Interface(
    fn=generate_recipe,
    inputs=gr.Textbox(label="Enter your prompt", placeholder="Example: To make fluffy pancakes, you need..."),
    outputs=gr.Textbox(label="Generated Recipe"),
    title="🍳 Recipe Generator",
    description="Custom fine-tuned GPT model that generates recipes from your prompt."
)

# Launch interface
demo.launch()
```
